```python
mp["age"] = (
    mp["Missing Age"]
      .str.replace(r"\s*years?\s*$", "", regex=True, case=False)
      .str.replace(r"\s*yrs?\s*$", "", regex=True, case=False)
      .str.strip()
)
```

```python
mp
```

```python
mp["age"] = pd.to_numeric(mp["age"], errors="coerce")
```

```python
bins = [0, 17, 34, 54, 150] 
labels = ["0-17", "18-34", "35-54", "55+"]
```

```python
mp["age"] = pd.cut(mp["age"], bins=bins, labels=labels, include_lowest=True)
```
