```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
pop_df=pd.read_csv('../data/Population data.csv')
```

```python
mp=pd.read_csv('../Notebook/My_Capstone.csv')
```

```python
mp
```

```python
pop_df = pop_df.dropna(axis=1, how='all')
```

```python
mp['County_2']= mp['County'] + ', ' + mp['State']
```

```python
mp.rename(columns={'County': 'County_old', 'County_2': 'County'}, inplace=True)
```

```python
mp
```

```python
mp['Date Last Contacted'].value_counts()
```

```python
mp['Missing_Rate'] = (mp['missing_person_count'] / mp['population']) * 100000
```

```python
mp
```

```python
tn_missing = mp[mp['State'] == 'TN']
```

```python
pop_melted = pop_df.melt(id_vars=['County'], 
                      var_name='Year', 
                      value_name='Population')
```

```python
pop_melted
```

```python
pop_melted['Year'] = pop_melted['Year'].astype(int)
```

```python
merged = pd.merge(
    tn_missing,
    pop_melted,
    on=['County', 'Year'],
    how='left'
)
```

```python
merged
```

```python
merged['Rate_per_100k'] = (merged['missing_person_count'] / merged['population']) * 100000
```

```python
merged
```

```python
tn_df= merged.drop(columns=['Population'])
```

```python
tn_df['Race'] = tn_df['Race'].replace('Black', 'African American')
```

```python
mp['Race'] = mp['Race'].replace('Black', 'African American')
```

```python
tn_df
```

```python
tn_df = tn_df[(tn_df['State'] == 'TN') & (tn_df['Year'].between(2020, 2024))]
```

```python
tn_df
```

```python
pop_df.to_csv('population_by_year.csv', index=False)
```

```python
tn_df.to_csv('tn_population_by_county.csv', index=False)
```
